```python
import pandas as pd

from ms.metaresearch.runner import run_selector
```

```python
import numpy as np
import random

seed = 42
np.random.seed(seed)
random.seed(seed)
```

```python
from ms.features_list import to_loc

features = pd.read_csv("../resources/tabzilla/filtered/features.csv", index_col=0)
features = features.loc[:, to_loc]
```

```python
metrics_raw = pd.read_csv("../resources/tabzilla/target/metrics__raw.csv", index_col=0)
```

```python
metrics_abs = pd.read_csv("../resources/tabzilla/target/metrics__perf_abs.csv", index_col=0)
```

```python
metrics_rel = pd.read_csv("../resources/tabzilla/target/metrics__perf_rel.csv", index_col=0)
```

```python
metrics_diff = pd.read_csv("../resources/tabzilla/target/metrics__diff.csv", index_col=0)
```

```python
datasets = features.merge(metrics_raw, on="dataset_name").index
features = features.loc[datasets]
metrics_raw = metrics_raw.loc[datasets]
```

```python
from sklearn.model_selection import KFold
from ms.metaresearch.data_split import split

outer_split = KFold(n_splits=3, shuffle=True, random_state=42)
inner_split = KFold(n_splits=5, shuffle=True, random_state=42)
data_splits = split(features, metrics_raw, outer_split, inner_split)
```

```python
from sklearn.linear_model import Lasso
from ms.metaresearch.selectors.model_based import LassoSelector, XGBSelector
from ms.metaresearch.selectors.model_free import MutualInfoSelector, FValueSelector
from ms.metaresearch.selectors.causal import CFSelector
from ms.metaresearch.selectors.model_free import CorrelationSelector

corr = CorrelationSelector(
    corr_type="spearman",
    p_threshold=0.1,
    corr_threshold=0.0,
)
mi = MutualInfoSelector(
    quantile_value=0.0,
)
f_value = FValueSelector(
    p_threshold=0.1,
    quantile_value=0.0,
)
lasso = LassoSelector(
    coef_threshold=0.0,
    reg_params=None,
    class_params=None,
    random_state=seed,
)
xgb = XGBSelector(
    importance_threshold=0.0,
    reg_params=None,
    class_params=None,
    random_state=seed,
)
# te = TESelector(
#     corr_selector=None,
#     to_tune=True,
#     model_y=Lasso(),
#     model_t=Lasso(),
#     quantile_value=0.8,
#     max_depth=50,
#     min_leaf=5,
#     n_splits=2,
#     n_estimators=100,
#     n_trees=500,
#     n_jobs=-1,
#     mode="individual",
#     model_type="CausalForestDML",
#     random_state=None,
# )

cf = CFSelector()
    
```

```python
from ms.metaresearch.selectors.causal import TEDAGSelector

te_dag = TEDAGSelector()
```

```python
metrics = {
    # "raw": metrics_raw, 
    "abs": metrics_abs, 
    "rel": metrics_rel, 
    "diff": metrics_diff
}
selectors = {
    "corr": corr,
    # "mi": mi,
    "f_val": f_value, 
    # "lasso": lasso,
    "xgb": xgb,
    # "te": te,
    "cf": cf,
    "te_dag": te_dag,
}
```

```python
from pathlib import Path
from ms.utils.navigation import pjoin, get_project_path

for metric_name, metric in metrics.items():
    print(f"Running {metric_name}...")
    save_path = Path(pjoin(get_project_path(), "results", "tabzilla", metric_name))
    for selector_name, selector in selectors.items():
        print(f"Running {selector_name}...")
        run_selector(
            selector,
            features,
            metric,
            data_splits,
            "quantile",
            k_best=10,
            save_path=pjoin(
                save_path, 
                selector.name, 
                "res.json"
            )
        )
```
